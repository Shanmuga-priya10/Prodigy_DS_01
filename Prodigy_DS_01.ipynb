```python
# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the World Bank Population Dataset
# Replace the filename with the correct CSV file extracted from the downloaded ZIP
df = pd.read_csv("/content/sample_data/API_SP.POP.TOTL_DS2_en_csv_v2_382940.csv", skiprows=4)

# Step 3: Preview the Dataset
print("Dataset Preview:")
print(df.head())

# Step 4: Select Data for the Year 2022
df_selected = df[["Country Name", "2022"]]   # You can change the year if needed
df_selected = df_selected.dropna()           # Remove rows with missing population values

# Step 5: Sort and Select Top 10 Most Populated Countries
top_10 = df_selected.sort_values(by="2022", ascending=False).head(10)

print("\nTop 10 Most Populated Countries in 2022:")
print(top_10)

# Step 6: Create a Bar Chart
plt.figure(figsize=(12, 7))
sns.barplot(x="2022", y="Country Name", data=top_10, palette="coolwarm")
plt.title("Top 10 Most Populated Countries in 2022", fontsize=16)
plt.xlabel("Population", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Step 7: Show and Save the Chart
plt.savefig("top_10_population_2022.png")
plt.show()
```
